```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
import math

x = Symbol('x')
y1=nsolve(0.0155+1.712*x-4.9*x**2-0.07*sin(25.1*x+0.22),x,1)
print("Solución a la ecuación trascendental")
print(y1)
R=0.0155+1.712*y1-4.9*y1**2
print("Posición en la que nuevamente se intersectan las dos masas")
print(R)
t = np.arange(0,0.4,0.001)   # start,stop,step
y = 0.0155+1.71*t-4.9*t**2
z = 0.07*np.sin(25.1*t+0.22)
fig = plt.figure()
plt.title("Posición Vs Tiempo ")
plt.xlabel("Tiempo [s]")
plt.ylabel("Posición [m]")
plt.plot(t,z,"r")
plt.plot(t,y,"b")
plt.grid()
plt.show()
```
